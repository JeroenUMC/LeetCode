```python
from typing import Dict, List, Optional, Tuple
```

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if not prices:
            return 0

        buy1 = float("-inf")
        sell1 = 0
        buy2 = float("-inf")
        sell2 = 0

        for price in prices:
            buy1 = max(buy1, -price)          # Best after first buy
            sell1 = max(sell1, buy1 + price)  # Best after first sell
            buy2 = max(buy2, sell1 - price)   # Best after second buy
            sell2 = max(sell2, buy2 + price)  # Best after second sell

        return sell2
```

```python
prices = [3,2,6,5,0,3]
sol = Solution()
print(sol.maxProfit(prices))
```

```python
# 20 test cases for maxProfit and state visualization
TEST_CASES_123 = [
    ("empty_prices", [], 0),
    ("single_day", [5], 0),
    ("two_days_descending", [5, 4], 0),
    ("two_days_ascending", [1, 2], 1),
    ("monotonic_increasing", [1, 2, 3, 4, 5], 4),
    ("monotonic_decreasing", [7, 6, 4, 3, 1], 0),
    ("leetcode_example", [3, 3, 5, 0, 0, 3, 1, 4], 6),
    ("two_good_windows", [1, 2, 4, 2, 5, 7, 2, 4, 9, 0], 13),
    ("small_two_transactions", [2, 1, 2, 0, 1], 2),
    ("staggered_peaks", [6, 1, 3, 2, 4, 7], 7),
    ("split_rise", [1, 4, 2, 7], 8),
    ("late_big_rise", [2, 4, 1, 7], 8),
    ("classic_high_profit", [5, 11, 3, 50, 60, 90], 93),
    ("all_equal", [2, 2, 2, 2], 0),
    ("three_opportunities_take_two", [1, 5, 1, 5, 1, 5], 8),
    ("alternating_high_low", [9, 1, 9, 1, 9], 16),
    ("deep_dip_then_peak", [4, 2, 8, 1, 9], 14),
    ("wave_with_final_rally", [1, 3, 2, 8, 4, 9], 12),
    ("mid_drop_second_trade", [3, 8, 5, 1, 7, 8], 12),
    ("textbook_two_trade", [10, 22, 5, 75, 65, 80], 87),
]


def trace_states(prices: List[int]):
    """Return per-day state transitions for the 2-transaction DP."""
    buy1 = float("-inf")
    sell1 = 0
    buy2 = float("-inf")
    sell2 = 0

    rows = []
    for day, price in enumerate(prices):
        buy1 = max(buy1, -price)
        sell1 = max(sell1, buy1 + price)
        buy2 = max(buy2, sell1 - price)
        sell2 = max(sell2, buy2 + price)
        rows.append((day, price, buy1, sell1, buy2, sell2))

    return rows, sell2


def print_state_table(rows):
    headers = ("day", "price", "buy1", "sell1", "buy2", "sell2")
    line = f"{headers[0]:>3} | {headers[1]:>5} | {headers[2]:>6} | {headers[3]:>6} | {headers[4]:>6} | {headers[5]:>6}"
    print(line)
    print("-" * len(line))
    for day, price, buy1, sell1, buy2, sell2 in rows:
        print(f"{day:>3} | {price:>5} | {buy1:>6.0f} | {sell1:>6.0f} | {buy2:>6.0f} | {sell2:>6.0f}")


all_ok = True
for i, (name, prices, expected) in enumerate(TEST_CASES_123, start=1):
    rows, computed = trace_states(prices)
    ok = computed == expected
    all_ok = all_ok and ok

    print(f"\nCase {i:02d} - {name}")
    print(f"prices   = {prices}")
    print(f"expected = {expected}, computed = {computed}, pass = {ok}")

    if rows:
        print_state_table(rows)
    else:
        print("(no days to process)")

print("\nSummary:", "all 20 cases passed" if all_ok else "some cases failed")
```
