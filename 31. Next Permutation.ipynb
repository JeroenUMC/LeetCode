```python
from typing import List
from itertools import permutations
```

```python
class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        n = len(nums)

        # 1. find pivot
        i = n - 2
        while i >= 0 and nums[i] >= nums[i + 1]:
            i -= 1

        if i == -1:
            nums.reverse()
            return   # last permutation, return first permutation

        # 2. find successor
        j = n - 1
        while nums[j] <= nums[i]:
            j -= 1

        # 3. swap
        nums[i], nums[j] = nums[j], nums[i]

        # 4. reverse suffix
        nums[i + 1:] = reversed(nums[i + 1:])
    
        return
```

```python
sol = Solution()
```

```python
arr = [3,2,1]

result = sol.nextPermutation(arr)

print(result)
```
