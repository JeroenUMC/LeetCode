```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x == 0 or x == 1:
            return x

        low, high = 1, x // 2
        
        while low <= high:
            mid = (low + high) // 2
            squared = mid * mid
            
            if squared == x:
                return mid
            elif squared < x:
                # Check if this is our answer (mid² ≤ x < (mid+1)²)
                if (mid + 1) * (mid + 1) > x:
                    return mid
                low = mid + 1
            else:
                high = mid - 1
        
        return high
```

```python
sol = Solution()
```

```python
x = 8192
result = sol.mySqrt(x)

print(result)
```
