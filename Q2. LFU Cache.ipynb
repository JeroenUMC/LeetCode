```python
from collections import OrderedDict
```

```python
class LFUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.cashe = [OrderedDict()]
    def get(self, key: int) -> int:
        # return key if exists, else -1
        for index,layer in enumerate(self.cashe):
            if key in layer:
                value = layer.pop(key)

                next_index = index + 1
                while next_index >= len(self.cashe):
                    self.cashe.append(OrderedDict())

                self.cashe[next_index][key] = value
                return value
        return -1
    
    def put(self, key: int, value: int) -> None:
        # update the value of the key if the key exists 
        # Otherwise add the key
        # if the key exceeds capacity, evict the least recently used key
        
        # Existing key: promote frequency and update value
        for index, layer in enumerate(self.cashe):
            if key in layer:
                layer.pop(key)
                target_index = index + 1
                while target_index >= len(self.cashe):
                    self.cashe.append(OrderedDict())
                self.cashe[target_index][key] = value
                return

        # New key: evict first if full
        total = sum(len(layer) for layer in self.cashe)
        if total >= self.capacity:
            for layer in self.cashe:
                if layer:
                    layer.popitem(last=False)
                    break

        # Insert new key at frequency 0
        if len(self.cashe) == 0:
            self.cashe.append(OrderedDict())
        self.cashe[0][key] = value
```

```python
# Test case: LFUCache sequence
operations = ["LFUCache", "put", "put", "get", "get", "get", "put", "put", "get", "get", "get", "get"]
arguments = [[3], [2, 2], [1, 1], [2], [1], [2], [3, 3], [4, 4], [3], [2], [1], [4]]
expected = [None, None, None, 2, 1, 2, None, None, -1, 2, 1, 4]

results = []
cache = None



for op, args in zip(operations, arguments):
    if op == "LFUCache":
        cache = LFUCache(args[0])
        results.append(None)
    elif op == "put":
        cache.put(args[0], args[1])
        results.append(None)
    elif op == "get":
        results.append(cache.get(args[0]))

print("Operations:", operations)
print("Arguments:", arguments)
print("Expected:", expected)
print("Actual:  ", results)
print("Pass:", results == expected)
```
